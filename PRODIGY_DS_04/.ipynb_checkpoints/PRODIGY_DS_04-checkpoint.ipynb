```python
# Analyze and visualize sentiment patterns in social media data to understand public opinion and attitudes towards specific topics or brands.
# Sample Dataset :- https://www.kaggle.com/datasets/jp797498e/twitter-entity-sentiment-analysis
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Twitter Sentiment Analysis.xlsx")
```

```python
df.head()
```

```python
len(df)
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(len(df))
df.head()
```

```python
import re

df['Description'] = df['Description'].apply(lambda x: re.sub(r'[^A-Za-z0-9 ]', '', str(x)) if pd.notnull(x) else '')
df['Description'] = df['Description'].apply(lambda x: re.sub(r'^\d+|\d+$', '', str(x)).strip())
df['Description'] = df['Description'].str.strip()
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df['Companies'].unique()
```

```python
df['Thoughts'].unique()
```

```python
sns.countplot(data=df,x='Thoughts')
plt.title('Distribution of Sentiments')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.countplot(data=df,x='Companies',hue='Thoughts')
plt.xticks(rotation=90) 
plt.title('Distribution of Sentiments according to the companies')
plt.show()
```

```python
piv = df.pivot_table(index='Companies', columns='Thoughts', aggfunc='size', fill_value=0)

plt.figure(figsize=(14, 7))
piv.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Sentiment Counts by Company')
plt.xticks(rotation=90)
plt.legend(title='Sentiment')
plt.show()
```
