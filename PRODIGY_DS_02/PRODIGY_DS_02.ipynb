```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
len(df)
```

```python
df.drop('deck',axis=1,inplace=True)
df.isnull().sum()
```

```python
# Embarked and Embarked town is same. So I am eliminating the column Embarked town
df.drop('embark_town',axis=1,inplace=True)
df.sample()
```

```python
df['embarked'].mode()[0]
```

```python
df['embarked'].fillna(df['embarked'].mode()[0],inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
(df['age'].isnull().sum()/len(df))*100
```

```python
print("Mean:",round(df['age'].mean()))
print("Medain:",round(df['age'].mean()))
print("Mode:",df['age'].mode())
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
sns.countplot(data=df, x='pclass', ax=axes[0, 0])
sns.countplot(data=df, x='sex', ax=axes[0, 1])
sns.countplot(data=df, x='embarked', ax=axes[1, 0])
sns.countplot(data=df, x='class', ax=axes[1, 1])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='fare', hue='survived', palette='coolwarm')
plt.title('Age vs Fare by Survival')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='class', y='survived')
plt.title('Survival Rate by Class')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='sex', y='survived')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='embarked', y='survived')
plt.title('Survival Rate by Embarkation Point')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df[['age', 'sibsp', 'parch', 'fare', 'survived']].corr(), annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df['family'] = df['sibsp'] + df['parch'] + 1

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='family', y='survived')
plt.title('Survival Rate by Family')
plt.show()
```
